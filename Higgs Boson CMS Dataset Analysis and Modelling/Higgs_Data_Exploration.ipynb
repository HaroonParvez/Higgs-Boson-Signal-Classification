```python
# Investigating fill values, I used a for loop to check for several all at once.
fill_values = (-999, -1, 0) # typical fill values
for value in fill_values:
    print(f"\nCount of {value} in each column:")
    print((df == value).sum()) # checks for the current fill value in the data
```

No use of -999.

Usage of -1 was found in tau_vertextEnergyRatio_0 and tau_vertextEnergyRatio_1. This is affecting the data set, as -1 is being used as a fill value for missing data. This will cause bias and influence the analysis if not handled.

Zero values were found in the nSV, isBackground and iSignal columns. These are NOT missing or fill values, and are being used as a classification. These will not need to be handled.

```python
# Replace -1 with NaN in the affected columns
df['tau_vertexEnergyRatio_0'].replace(-1, np.nan, inplace=True)
df['tau_vertexEnergyRatio_1'].replace(-1, np.nan, inplace=True)

# Drop rows where there is NaN
df.dropna(subset=['tau_vertexEnergyRatio_0', 'tau_vertexEnergyRatio_1'], inplace=True)

# Check new shape of dataset
print(df.shape)
```

We have lost a very tiny amount of data (102 rows), this should not have much affect on analysis.

```python
# Create histograms comparing background and signal labels for each numerical variable
# List of features for plotting
features = df.columns.drop(['Unnamed: 0', 'isBackground', 'isSignal'])
```

For bin size I will be using 50 for all histograms, I believe this is a well balanced size allowing for easy readability of the histograms, and anything odd to be noticed easily. It is not too small and not too big/noisy. 

```python
# Plotting histograms
for feature in features:
    # Set x-axis label (to add units where applicable)
    if feature in ['tau_vertexMass_0', 'tau_vertexMass_1']:
        xlabel = f'{feature} (GeV/c²)'
    else:
        xlabel = feature
    # Set y-axis label for linear scale
    if feature in ['tau_vertexMass_0', 'tau_vertexMass_1']:
        ylabel_linear = 'Density (1/(GeV/c²))'
    else:
        ylabel_linear = 'Density'
    # Create a new figure for each feature (linear)
    plt.figure(figsize=(8,6))
     # Plot histogram for background
    plt.hist(df[df['isBackground'] == 1][feature], bins=50, density=True, alpha=0.5, label='Background')
    # Plot histogram for signal
    plt.hist(df[df['isSignal'] == 1][feature], bins=50, density=True, alpha=0.5, label='Signal')
    # Add labels and legend
    plt.xlabel(feature, fontsize=14)
    plt.ylabel('Density', fontsize=14)
    plt.title(f'{feature} Distribution (Linear Scale)', fontsize=16)
    plt.legend()
    plt.show()
```

```python
# Plotting log scale Histograms
for feature in features:
    # Set x-axis label (add unit if applicable)
    if feature in ['tau_vertexMass_0', 'tau_vertexMass_1']:
        xlabel = f'{feature} (GeV/c²)'
    else:
        xlabel = feature
    # Create a new figure for log scale
    plt.figure(figsize=(8,6))
    # Plot histogram for background
    plt.hist(df[df['isBackground'] == 1][feature], bins=50, density=True, alpha=0.5, label='Background')

    # Plot histogram for signal
    plt.hist(df[df['isSignal'] == 1][feature], bins=50, density=True, alpha=0.5, label='Signal')

    # Set y-axis to logarithmic scale
    plt.yscale('log')

    # Add labels and legend
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel('Density (log scale)', fontsize=14)
    plt.title(f'{feature} Distribution (Log Scale)', fontsize=16)
    plt.legend()
    # Show plot
    plt.show()
```

**Advantages of Linear Scale:**
- The relative heights of different regions of the distribution are clearly visible.
- Peaks and general shapes of distributions are easy to interpret for regions where the event density is high.

**Disadvantages of Linear Scale:**
- If a distribution has a long tail or events spread over many orders of magnitude, the lower-count regions may not be visible at all.
- Rare but important differences between background and signal may be hidden at times.

**Advantages of Logarithmic Scale:**
- Highlights the structure of the distributions across several orders of magnitude.
- Rare events or small differences between background and signal can be clearly seen.
- Allows better visibility of tails or outliers in the data.

**Disadvantages of Logarithmic Scale:**
- Compresses the main peak regions, making it harder to see small differences there.
- Log scale can exaggerate fluctuations in lower count regions.

**Conclusion:**
Both linear and logarithmic scales provide useful insights.  
Linear plots are better for visualising the general shape and bulk of the distribution, while logarithmic plots are better for exploring the tails and small-probability regions.  
Using both types of plots together ensures that the full structure of the data is analysed and understood.

```python
# Calculating mean and standard deviation for each numerical variable
# First need to exclude index and label columns:
features = df.columns.drop(['Unnamed: 0', 'isBackground', 'isSignal'])

# Create a data frame to store the results
mean_std_table = pd.DataFrame({
    'Mean': df[features].mean(),
    'Standard Deviation': df[features].std()})

# Display the table
mean_std_table
```

A table is good for this as it keeps it simple and allows for easy comparison. I will also plot it below for a better visual comparison.

```python
# Plotting means and standard deviations side by side
plt.figure(figsize=(14, 6))
x = np.arange(len(features))  # features and
width = 0.4  # width of the bars

# One half of the bar will be mean, other half will be standard deviation
plt.bar(x - width/2, mean_std_table['Mean'], width=width, label='Mean')
plt.bar(x + width/2, mean_std_table['Standard Deviation'], width=width, label='Std Dev')

plt.xticks(ticks=x, labels=features, rotation=90)
plt.ylabel('Value')
plt.title('Mean and Standard Deviation of Features')
plt.legend()
plt.grid(True, axis='y')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df[features].corr()

# Making a table of the correlation coefficents
correlation_table = correlation_matrix

# Display the table
print("Correlation Coefficients Table")
display(correlation_table)

# Plot the heat map
plt.figure(figsize=(12, 10)) # Size must be large to accommodate all of it nicely
plt.imshow(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(label='Correlation Coefficient')
plt.xticks(ticks=np.arange(len(features)), labels=features, rotation=90)
plt.yticks(ticks=np.arange(len(features)), labels=features)
plt.title('Correlation Heatmap', fontsize=16)
plt.grid(False)
plt.show()
```

```python
# Find the most positively and negatively correlated variable pairs
corr_pairs = correlation_matrix.unstack()
# Remove self-correlations (just to make it easier)
corr_pairs = corr_pairs[corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)]

# Sort correlations
sorted_corr = corr_pairs.sort_values()

# Display top 2 most negative and top 2 most positive correlations
print("Most Negatively Correlated Pair")
print(sorted_corr.head(4))
print()
print("Top 2 Most Positively Correlated Pairs:")
print(sorted_corr.tail(4))
```

**The first two for both ends are the same pair in a different, so I produced the top 4 for the top end and lower end to show the top 2 unique pairs for most positive and most negative.**

```python
# List of the top 2 positive/negative variable pairs
pairs_to_plot = [
    ('tau0_trackEtaRel_1', 'tau_vertexDeltaR_0'),   # Most negative pair 1
    ('tau0_trackEtaRel_0', 'tau_vertexDeltaR_0'),   # Most negative pair 2
    ('tau1_trackEtaRel_1', 'tau1_trackEtaRel_2'),   # Most positive pair 1
    ('trackSip2dSigAboveBottom_1', 'trackSip2dSigAboveBottom_0')]  # Most positive pair 2

# Plot 2D histograms for each pair
for var1, var2 in pairs_to_plot:
    plt.figure(figsize=(8,6))
    plt.hist2d(df[var1], df[var2], bins=50)
    plt.xlabel(var1, fontsize=14)
    plt.ylabel(var2, fontsize=14)
    plt.title(f'2D Histogram: {var1} vs {var2}', fontsize=16)
    plt.colorbar(label='Counts')
    plt.show()
```

Most positively correlated pairs show a strong clustering along the diagonal, indicating that as one variable increases, the other increases in a consistent manner.

Most negatively correlated pairs show an inverse pattern, where an increase in one variable corresponds to a decrease in the other, forming a downward trend.

These observations confirm the correlation values calculated and highlight strong relationships between these features.
